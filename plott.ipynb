```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("supervised_employee_feedback.csv")

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sentiment', hue='Sentiment', palette='Set2', legend=False)

plt.title("Count of Feedbacks by Sentiment", fontsize=16)
plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.savefig("sentiment_count_plot.png")
plt.show()

```

## Sentiment Distribution â€“ Count Plot

This bar chart displays the number of feedback entries classified under each sentiment category: **Positive**, **Neutral**, and **Negative**.

It helps us quickly understand which sentiment is most common in the employee feedback dataset.

- X-axis: Sentiment categories  
- Y-axis: Number of feedback entries  
- Insight: A higher count in a particular sentiment indicates its dominance in the dataset.


```python
sentiment_counts = df['Sentiment'].value_counts()
colors = ['lightgreen', 'skyblue', 'coral']

plt.figure(figsize=(6, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Sentiment Distribution", fontsize=14)
plt.tight_layout()
plt.savefig("sentiment_pie_chart.png")
plt.show()
```

## Sentiment Distribution â€“ Pie Chart

The pie chart shows the **percentage share of each sentiment class** in the dataset.

This is useful for getting a proportion-based understanding of the sentiment distribution across the entire dataset.

- Each slice represents the share of feedbacks that are Positive, Neutral, or Negative  
- The values are expressed in percentages  
- Helps identify imbalance or skew in the sentiment classes visually.
